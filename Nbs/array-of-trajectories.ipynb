```python
src0 = os.path.join(rootdir, 'Spontaneous_edge_tracking')
dst0 = os.path.join(rootdir, 'jyr', 'vertical_strip', 'Spontaneous_edge_tracking')
if not os.path.exists(dst0):
    os.makedirs(dst0)
```

```python
copy_files=False
if copy_files:
    import shutil
    src_files = glob.glob(os.path.join(src0, '*.log'))
    if not os.path.exists(os.path.join(dst0, 'logfiles')):
        os.makedirs(os.path.join(dst0, 'logfiles'))
    for sfile in src_files:
        fn = os.path.split(sfile)[-1]
        dfile = os.path.join(dst0, 'logfiles', fn)
        shutil.copy(sfile, dfile)
```

```python
srcdir = dst0
all_logfiles = glob.glob(os.path.join(srcdir, '*', '*.log'))
len(all_logfiles)

figid = srcdir.split('/My Drive')[-1]
print(figid)
```

```python
create_new=False
combined_datafile = os.path.join(srcdir, 'combined.pkl')
if not create_new:
    try:
        with open(combined_datafile, 'rb') as f:
            df0_all = pkl.load(f)
        print(df0_all.shape)
    except Exception as e:
        print(e)
        create_new=True
print(create_new)
```

```python
create_new=True
if create_new:
    d_list = []
    params = {}
    for fp in all_logfiles:
        parent, fn = os.path.split(fp)
        fname = os.path.splitext(fn)[0]

        curr_figdir = os.path.join(parent, 'trajectories')
        if not os.path.exists(curr_figdir):
            os.makedirs(curr_figdir)

        currcond = os.path.split(parent)[-1]
        df0_full, oparams = traj.load_df(fp)
        odor_ix = oparams['odor_ix']
        odor_bounds = oparams['odor_bounds']

        df0_full['condition'] = currcond
        params[fname] = oparams

        fig, ax = pl.subplots()
        ax = butil.plot_trajectory(df0_full, ax=ax, odor_bounds=odor_bounds, 
                             zero_odor_start=True)
        ax.set_title(fn, fontsize=8, loc='left')
        ax.set_aspect(1)

        util.label_figure(fig, '{}'.format(parent.split('/My Drive')[-1]))
        pl.savefig(os.path.join(curr_figdir, '{}.png'.format(fname)), dpi=300)
        pl.close()

        d_list.append(df0_full)
    # combine
    df0_all = pd.concat(d_list, axis=0).reset_index()
    # save
    with open(combined_datafile, 'wb') as f:
        pkl.dump(df0_all, f)
        
create_new=False
```

```python
nr=2
nc=6
fig, axn = pl.subplots(nr, nc, figsize=(11,8), sharex=True)
for fi, (fn, df_) in enumerate(df0_all.groupby('filename')):
    if fi >= nr*nc:
        break
    ax=axn.flat[fi]
    odor_ix = df_[df_['instrip']].iloc[0].name
    #plotdf = df_.loc[odor_ix:]
    # odor_ix = params[fn]['odor_ix']
    plotdf = df_.copy()
    offset_x = plotdf[plotdf['instrip']].iloc[0]['ft_posx']
    offset_y = plotdf[plotdf['instrip']].iloc[0]['ft_posy']
    plotdf['ft_posx'] = plotdf['ft_posx'].values - offset_x
    plotdf['ft_posy'] = plotdf['ft_posy'].values - offset_y
    odor_bounds = butil.find_strip_borders(plotdf, entry_ix=odor_ix,
                                        strip_width=50, #params[fn]['strip_width'], 
                                        strip_sep=params[fn]['strip_sep'])
    # plot
    plotdf = plotdf.loc[odor_ix:].copy()
    
    ax.plot(plotdf['ft_posx'], plotdf['ft_posy'], lw=1.5, c='w')
    for ob in odor_bounds:
        butil.plot_odor_corridor(ax, odor_xmin=ob[0], 
                             odor_xmax=ob[1], odor_linewidth=1)
    for bnum, b_ in plotdf[plotdf['instrip']].groupby('boutnum'):
        ax.plot(b_['ft_posx'], b_['ft_posy'], lw=1.5, c='r')
    # legend
    ax.axis('off')
    if fi==0:
        leg_xpos=-150; leg_ypos=0; leg_scale=100
        putil.vertical_scalebar(ax, leg_xpos=leg_xpos, leg_ypos=leg_ypos)
    #ax.set_box_aspect(3)
    ax.set_xlim([-150, 150])
    ax.set_ylim([-100, 1600])

pl.tight_layout()
pl.subplots_adjust(top=0.85, hspace=0.4, wspace=0.5) #left=0.1, right=0.9, wspace=1, hspace=1, bottom=0.1, top=0.8)

util.label_figure(fig, figid)

figname = '{}-traj-together'.format(nr*nc)
pl.savefig(os.path.join(curr_figdir, '{}.png'.format(figname)))
```
